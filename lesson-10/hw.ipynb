```python
from sklearn.datasets import load_digits
df=load_digits()
```

```python
x=df.data
y=df.target
```

```python
x.shape
```

```python
import numpy as np
np.unique(y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.matshow(df.images[2], cmap='gray')
plt.colorbar()
plt.show()
```

```python
print('Min pixel:', x.min())
print('Max pixel:', x.max())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(max_iter=1000)
log_reg.fit(x_train, y_train)
y_pred_logreg=log_reg.predict(x_test)
```

```python
y_pred_logreg
```

```python
from sklearn.svm import SVC
svm_clf=SVC(probability=True)
svm_clf.fit(x_train, y_train)
y_pred_svm=svm_clf.predict(x_test)
```

```python
y_pred_svm
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
print('Logistic Regression Accuracy:', accuracy_score(y_test, y_pred_logreg))
print('Classification Report:\n', classification_report(y_test, y_pred_logreg))
cm_logreg=confusion_matrix(y_test, y_pred_logreg)
plt.figure(figsize=(10,7))
sns.heatmap(cm_logreg, annot=True, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

```

```python
print('SVM Accuracy:', accuracy_score(y_test, y_pred_svm))
print('Classification Report:\n', classification_report(y_test, y_pred_svm))
cm_svm=confusion_matrix(y_test, y_pred_svm)
plt.figure(figsize=(10,7))
sns.heatmap(cm_svm, annot=True, cmap='Greens')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
models={'Logistic Regression':log_reg, 'SVM': svm_clf}
for name, model in models.items():
    print(f'{name} Metrics')
    print('Macro Precision:', precision_score(y_test, model.predict(x_test), average='macro'))
    print('Macro Recall:', recall_score(y_test, model.predict(x_test), average='macro'))
    print('Macro F1-Score:', f1_score(y_test, model.predict(x_test), average='macro'))
    print('Weighted Precision:', precision_score(y_test, model.predict(x_test), average='weighted'))
    print('Weighted Recall:', recall_score(y_test, model.predict(x_test), average='weighted'))
    print('Weighted F1-Score:', f1_score(y_test, model.predict(x_test), average='weighted')) 
    print('---------------------------')    
```

```python
from sklearn.metrics import roc_curve, auc
y_score_logreg=log_reg.predict_proba(x_test)
y_score_svm=svm_clf.predict_proba(x_test)
```

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
y_test_binarized=lb.fit_transform(y_test)
```

```python
plt.figure(figsize=(10,7))
for i in range(len(lb.classes_)):
    fpr, tpr, _=roc_curve(y_test_binarized[:, i], y_score_logreg[:, i])
    roc_auc=auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Digit {i} (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
for i in range(len(lb.classes_)):
    fpr, tpr, _=roc_curve(y_test_binarized[:, i], y_score_svm[:, i])
    roc_auc=auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Digit {i} (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.legend()
plt.show()
```
