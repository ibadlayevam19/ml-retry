```python
from sklearn.datasets import load_wine
data=load_wine()
```

```python
x=data.data
y=data.target
```

```python
columns=data.feature_names
target=data.target_names
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(x, columns=columns)
df['target']=y
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
lr=LogisticRegression(random_state=42, max_iter=2000)
lr.fit(x_train_scaled, y_train)
lr_pred=lr.predict(x_test_scaled)
```

```python
dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_train_scaled, y_train)
dt_pred=dt.predict(x_test)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)
knn_pred=knn.predict(x_test_scaled)
```

```python
lr.score(x_test_scaled, y_test)
```

```python
knn.score(x_test_scaled, y_test)
```

```python
dt.score(x_test_scaled, y_test)
```

```python
from sklearn.ensemble import VotingClassifier
hard_voting=VotingClassifier(
    estimators=[
        ('lr', lr),
        ('knn', knn),
        ('dt', dt)
    ],
    voting='hard'
)
hard_voting.fit(x_train_scaled, y_train)
y_pred_hard=hard_voting.predict(x_test_scaled)
hard_voting.score(x_test_scaled, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred_hard, y_test)
```

```python
soft_voting=VotingClassifier(
    estimators=[
        ('lr', lr),
        ('knn', knn),
        ('dt', dt)
    ],
    voting='soft'
)
soft_voting.fit(x_train_scaled, y_train)
y_pred_soft=soft_voting.predict(x_test_scaled)
soft_voting.score(x_test_scaled, y_test)
```
